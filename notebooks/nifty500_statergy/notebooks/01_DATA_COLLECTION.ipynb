```python
import pandas as pd

constituents = pd.read_csv(
    "../data/ind_nifty500list.csv"
)

print(constituents.head())

print("\nColumns:")
print(constituents.columns)

print("\nNumber of Stocks:")
print(len(constituents))
```

```python
symbols = (
    constituents["Symbol"]
    + ".NS"
).tolist()

print(symbols[:20])

print()

print("Total Symbols:", len(symbols))
```

```python
pd.Series(symbols).to_csv(
    "../data/nifty500_symbols.csv",
    index=False
)
```

```python
import yfinance as yf

test_symbols = symbols[:10]

test_data = yf.download(
    test_symbols,
    start="2015-01-01",
    end="2026-01-01",
    auto_adjust=True,
    progress=True
)

print(test_data.shape)
```

```python
import yfinance as yf

data = yf.download(
    symbols,
    start="2015-01-01",
    end="2026-01-01",
    auto_adjust=True,
    progress=True,
    group_by="ticker",
    threads=True
)
```

```python
data.to_pickle(
    "../data/nifty500_ohlcv.pkl"
)

print("Saved")
```

```python
close = pd.DataFrame()

for symbol in symbols:

    try:
        close[symbol] = data[symbol]["Close"]

    except:
        pass

print(close.shape)

close.head()
```

```python
close = pd.DataFrame()

for symbol in symbols:

    try:
        close[symbol] = data[symbol]["Close"]

    except:
        pass

print(close.shape)

close.head()
```

```python
print(close.shape)

print(close.index.min())
print(close.index.max())

print(close.isna().sum().sort_values(ascending=False).head(20))
```

```python
missing_pct = close.isna().mean()

close_clean = close.loc[
    :,
    missing_pct < 0.20
]

print("Original :", close.shape)
print("Cleaned  :", close_clean.shape)
```

```python
close_clean = close_clean.ffill()

close_clean.to_csv(
    "../data/nifty500_close_clean.csv"
)

close_clean.to_pickle(
    "../data/nifty500_close_clean.pkl"
)
```

```python
import pandas as pd

open_df = pd.DataFrame()
high_df = pd.DataFrame()
low_df = pd.DataFrame()
volume_df = pd.DataFrame()

for symbol in symbols:

    try:
        open_df[symbol] = data[symbol]["Open"]
        high_df[symbol] = data[symbol]["High"]
        low_df[symbol] = data[symbol]["Low"]
        volume_df[symbol] = data[symbol]["Volume"]

    except:
        pass

print("Open  :", open_df.shape)
print("High  :", high_df.shape)
print("Low   :", low_df.shape)
print("Volume:", volume_df.shape)
```

```python
open_df.to_pickle("../data/open.pkl")
high_df.to_pickle("../data/high.pkl")
low_df.to_pickle("../data/low.pkl")
volume_df.to_pickle("../data/volume.pkl")

print("Saved Successfully")
```
